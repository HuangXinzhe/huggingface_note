# 构建模型卡片
- 模型卡通常以非常简短的概述开始，说明模型的用途
- 模型描述
    - 提供了有关模型的基本详细信息。
        - 架构
        - 版本
        - 如果它是在论文中介绍的，是否有原始的实现可用？
        - 作者以及有关模型的一般信息
        - 任何版权都应归于此处。
    - 有关训练程序、参数和重要免责声明的一般信息
- 预期用途和限制
    - 在此描述模型可以适用的例子
        - 包括可以应用它的语言、领域。
    - 记录已知超出模型范围的区域，或者可能表现不佳的区域。
- 如何使用
    - 包括一些有关如何使用模型的示例。这可以展示使用 pipeline() 函数、模型和标记器类的使用以及其他任何您认为可能有帮助的代码。
- 局限性和偏见
- 训练数据
    - 指出模型是在哪个数据集上训练的。也欢迎对数据集进行简要描述。
- 训练程序
    - 描述从再现性角度来看有用的训练的所有相关方面。这包括对数据进行的任何预处理和后处理，以及模型训练的批量数、批量大小、学习率等细节。
- 评价结果
    - 应该提前在前面的部分告知，例如预期的使用效果和示例。最后，提供模型在评估数据集上的表现的指示。如果模型使用决策阈值，要么提供评估中使用的决策阈值，要么提供在不同阈值下针对预期用途进行评估的详细信息。

发布模型时不需要模型卡，制作一个模型时不需要包含上述所有部分。但是，模型的文档会使未来的用户受益，因此我们建议您尽自己的知识和能力填写尽可能多的部分。
